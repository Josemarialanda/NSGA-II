```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
import re
plt.style.use("ggplot")
url = "https://drive.google.com/file/d/14OPQj8e5smWiW0_HE9NhszAFDY43DFs5/view?usp=sharing"
url = "https://drive.google.com/uc?id=" + url.split("/") [-2]
df = pd.read_csv(url)
df.head()
```

```python
def extract_numeric_value(observation):
  return re.sub(r"\D", "", observation)

df["mileage"]= df["mileage"].apply(extract_numeric_value)
df.head()
```

```python
def extract_numeric_value_CM3(observation):
  return re.sub(r"\s*cm3", "", observation)

df["formatted_engine_capacity"]= df["engine_capacity"].apply(extract_numeric_value_CM3)
df.head()
```

```python
selected_columns = ["engine_capacity", "formatted_engine_capacity"]
selected_rows = 11

check_variable_creation = df[selected_columns].head(selected_rows)

print(check_variable_creation)
```

```python
df = df.drop(["engine_capacity"],axis=1)
df.head()
```
